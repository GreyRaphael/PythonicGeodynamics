```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
mpl.rcParams['figure.dpi']=100
```

```python
t_min, t_max = 0, 2*np.pi
intervals = 20
dt = (t_max -  t_min)/intervals
Nt = intervals + 1
time = np.linspace(t_min, t_max, Nt)
```

```python
x, y = np.cos(time), np.sin(time)
```

```python
vx = (x[1:] - x[:Nt-1])/dt
vx_avg = 0.5*(vx[:Nt-2] + vx[1:])
vy = (y[1:] - y[:Nt-1])/dt
vy_avg = 0.5*(vy[:Nt-2] + vy[1:])

plt.plot(time[1:Nt], vx, label='vx')
plt.plot(time[1:Nt-1], vx_avg, label='centered vx')
plt.plot(time, -np.sin(time), label='real vx')
plt.legend()
```

```python
ax=(vx_avg[1:]-vx_avg[:Nt-3])/dt
plt.plot(time[1:Nt-2], ax, label='ax(t)')
plt.plot(time, -np.cos(time),'o' ,label='real ax')
plt.legend()
```
