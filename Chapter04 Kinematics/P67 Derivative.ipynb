```python
import numpy as np
import matplotlib.pyplot as plt 
import matplotlib as mpl
```

```python
mpl.rcParams['figure.dpi']=100
```

```python
t_min, t_max = 0.0, 1.0
intervals = 20
dt = (t_max - t_min) / intervals
Nt = intervals+1
time=np.linspace(t_min, t_max, Nt)
```

```python
x=np.sin(2*np.pi*time)
```

```python
dxdt_backward = np.zeros(Nt)
dxdt_forward = np.zeros(Nt)
dxdt_centered = np.zeros(Nt)
```

```python
dxdt_backward[1:] = (x[1:] - x[:Nt-1])/ dt
dxdt_forward[:Nt-1] = (x[1:] - x[:Nt-1])/ dt
dxdt_centered[1:Nt-1]=0.5*(dxdt_backward[1:Nt-1] + dxdt_forward[1:Nt-1])
```

```python
plt.plot(time, x, label='x(t)')
plt.plot(time[1:], dxdt_backward[1:], label='back v(t)')
plt.plot(time[:Nt-1], dxdt_forward[:Nt-1], label='forward v(t)')
plt.plot(time[1:Nt-1], dxdt_centered[1:Nt-1], label='center v(t)')
plt.plot(time, 2*np.pi*np.cos(2*np.pi*time), 'o', label='real v(t)')
plt.xlabel('t')
plt.legend(loc='upper center')
```

```python
d2xdt2 = np.zeros(Nt)
d2xdt2[1:Nt-1] = (x[2:] - 2*x[1:Nt-1] + x[:Nt-2])/dt**2
```

```python
plt.plot(time, x, label='x(t)')
plt.plot(time[1:Nt-1], dxdt_centered[1:Nt-1], label='center v(t)')
plt.plot(time[1:Nt-1], d2xdt2[1:Nt-1], label='a(t)')
plt.plot(time, -4*np.pi**2*np.sin(2*np.pi*time),'o', label='readl a(t)')
plt.xlabel('t')
plt.legend()
```
