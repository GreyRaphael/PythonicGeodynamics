```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
mpl.rcParams['figure.dpi']=100
```

```python
t_min, t_max = 0.0, 4.0
intervals = 25
dt = (t_max - t_min) / intervals
Nt = intervals+1
time=np.linspace(t_min, t_max, Nt)
```

```python
acc = -10*np.ones(Nt)

vel = np.zeros(Nt)
vel[0] = 5

y = np.zeros(Nt)
y[0]=0
```

```python
# 没有forward, 只有backward
for i in range(Nt-1):
    vel[i+1]=vel[i]+acc[i]*dt
```

```python
for i in range(Nt-1):
    y[i+1]=y[i]+vel[i]*dt
```

```python
vel_avg=np.zeros(Nt)
vel_avg=0.5*(vel[:Nt-1] + vel[1:]) # len(vel_avg)==20
# vel是准确的，所以y_avg是准确的

y_avg = np.zeros(Nt)
y_avg[0] = 0

for i in range(Nt-1):
    y_avg[i+1]=y_avg[i]+vel_avg[i]*dt
```

```python
# plt.plot(time, acc, label='g(t)')

plt.plot(time, vel, label='v(t) backward')
plt.plot(time, y, label='y(t) backward')

plt.plot(time[:Nt-1], vel_avg, label='v(t) centered')
plt.plot(time, y_avg, label='y(t) centered')

plt.plot(time, acc*time+5, 'o', label='real v(t)')
plt.plot(time, 0.5*acc*time**2+5*time, 'o', label='real y(t)')
plt.legend()
```
