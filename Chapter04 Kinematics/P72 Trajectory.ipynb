```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
mpl.rcParams['figure.dpi'] = 100
```

```python
t_min, t_max = 0.0, 2.0
intervals = 25
dt = (t_max - t_min) / intervals
Nt = intervals+1
time=np.linspace(t_min, t_max, Nt)
```

```python
theta = np.pi/6
v0 = 8.0
gx, gy = 0.0, -10.0
ax, ay = np.ones(Nt)*gx, np.ones(Nt)*gy
vx, vy = np.zeros(Nt), np.zeros(Nt)
vx[0], vy[0] = v0*np.cos(theta), v0*np.sin(theta)
```

```python
for i in range(Nt-1):
    vx[i+1] = vx[i] + ax[i]*dt
    vy[i+1] = vy[i] + ay[i]*dt
```

```python
px, py = np.zeros(Nt), np.zeros(Nt)
vx_avg = 0.5*(vx[1:] + vx[:Nt-1])
vy_avg = 0.5*(vy[1:] + vy[:Nt-1])

for i in range(Nt-1):
    px[i+1] = px[i] + vx_avg[i]*dt
    py[i+1] = py[i] + vy_avg[i]*dt
```

```python
px_real = v0*np.cos(theta)*time
py_real = 0.5*ay*time**2 + v0*np.sin(theta)*time
```

```python
plt.plot(px_real, py_real, 'o', label='real trajectory')
plt.plot(px, py, label='centered trajectory')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.gca().set_aspect('equal')
```

- 如果速度的numerical是正好符合的，那么位移的numerical就会偏差;
- 如果位移的numerical是正好符合的, 那么速度的numerical就会偏差;
