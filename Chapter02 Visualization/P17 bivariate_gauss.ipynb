```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import matplotlib as mpl
```

```python
mpl.rcParams['figure.dpi']=100
```

```python
x, y=np.mgrid[-5:5:0.1, -5:5:0.1]
```

```python
# stack是new-axis; concatenate是existing-axis
pos=np.stack((x, y), axis=2)
```

[multivariate normal distribution](https://en.wikipedia.org/wiki/Multivariate_normal_distribution)

$$
\mu _{x,y}=\left[ \begin{array}{c}
	\mu _x\\
	\mu _y\\
\end{array} \right] ,\ cov_{x,y}=\left[ \begin{matrix}
	\sigma _{x}^{2}&		\rho \sigma _x\sigma _y\\
	\rho \sigma _x\sigma _y&		\sigma _{y}^{2}\\
\end{matrix} \right] 
$$


```python
rv1=stats.multivariate_normal(mean=[-2, -2], cov=[[1, 0], [0, 4]])
rv2=stats.multivariate_normal(mean=[1, 1], cov=[[4, 0], [0, 1]])
```

```python
z1=rv1.pdf(pos)
z2=rv2.pdf(pos)
z3=z1-z2
```

```python
plt.subplot(131)
p1=plt.contourf(x, y, z1)
plt.gca().set_aspect('equal')

plt.subplot(132)
p2=plt.contourf(x, y, z2)
plt.gca().set_aspect('equal')

plt.subplot(133)
p3=plt.contourf(x, y, z3)
plt.gca().set_aspect('equal')

plt.colorbar()
```

```python
levl=np.arange(-0.08, 0.1, 0.02)
plt.contourf(x, y, z3, levels=levl)
plt.colorbar()

# add contour line
cl=plt.contour(x, y, z3, levels=levl, colors='k')
# add contour line label
plt.clabel(cl, colors='k', fontsize=12)

# save fig to vector image
plt.savefig('p18.pdf')
```
